# 2. Introdução à Machine Learning (Classificação)

Os algoritmos de classificação são uma categoria fundamental de técnicas de aprendizado de máquina. Eles têm o propósito de atribuir rótulos ou categorias a dados com base em suas características. Esses algoritmos desempenham um papel crucial em uma ampla gama de aplicações, desde filtragem de spam de e-mails até análise de sentimentos em redes sociais e diagnóstico médico.

Alguns dos principais algoritmos de classificação são:

- **Naive Bayes:** Este é um algoritmo probabilístico que se baseia no Teorema de Bayes. É amplamente utilizado em classificação de textos, detecção de spam de e-mails e análise de sentimentos.

- **Regressão Logística:** Embora seja frequentemente usado para problemas de regressão, a regressão logística também é uma escolha popular para classificação binária. Ela modela a probabilidade de um exemplo pertencer a uma classe.

- **Árvores de Decisão:** As árvores de decisão dividem o conjunto de dados em subconjuntos menores com base em regras de decisão. São fáceis de interpretar, mas podem ser suscetíveis a overfitting.

- **Random Forest:** Uma extensão das árvores de decisão, o Random Forest constrói várias árvores e faz a média ou votação de suas previsões para obter uma classificação final. Isso ajuda a reduzir o overfitting.

- **Support Vector Machines (SVM):** O SVM é eficaz em problemas de classificação binária. Ele encontra um hiperplano que melhor separa as classes no espaço de características.

- **k-Nearest Neighbors (k-NN):** O k-NN classifica exemplos com base na classe predominante entre seus k vizinhos mais próximos no espaço de características.

- **Redes Neurais Artificiais:** Redes neurais profundas (Deep Learning), redes neurais convolucionais (CNN) e redes neurais recorrentes (RNN), têm sido usadas com sucesso em tarefas de classificação, especialmente em visão computacional e processamento de linguagem natural.

### 2.2 Treinamento

Nessa etapa vamos apresentar os dados de treinamento para o algoritmo escolhido (DecisionTreeClassifier).

```python
# Importa a biblioteca nltk (Natural Language Toolkit) para processamento de texto.
import nltk
# Baixa os dados necessários para o tokenizer, que divide o texto em palavras.
nltk.download('punkt')

# Importa o classificador DecisionTreeClassifier da scikit-learn
from sklearn.tree import DecisionTreeClassifier

# Importa a biblioteca para vetorização
from sklearn.feature_extraction.text import CountVectorizer

# Prepara os dados de treinamento e teste
X_train = train_df['Frase']
y_train = train_df['Sentimento']
X_test = test_df['Frase']
y_test = test_df['Sentimento']

# Vetoriza o texto usando a contagem de palavras (CountVectorizer)
vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Cria o modelo de árvore de decisão
clf = DecisionTreeClassifier(random_state=42)

# Treina o modelo
clf.fit(X_train_vectorized, y_train)
```

### 2.3 Avaliação

A matriz de confusão é uma ferramenta fundamental na avaliação de modelos de classificação. Ela fornece uma visão detalhada do desempenho do modelo, mostrando como suas previsões se comparam com os valores reais do conjunto de dados. A matriz de confusão é geralmente organizada em quatro quadrantes, e cada elemento da matriz representa uma contagem de instâncias:

- **Verdadeiros Positivos (True Positives - TP):** São os casos em que o modelo fez uma previsão correta, identificando corretamente uma instância como pertencente à classe positiva, e essa instância realmente pertence à classe positiva. Por exemplo, quando o modelo previu que a frase é do rotulo "gostei" e de fato a frase pertence a essa categoria.

- **Falsos Negativos (False Negatives - FN):** São os casos em que o modelo errou ao prever que uma instância pertence à classe negativa, quando na verdade a instância pertence à classe positiva. Por exemplo, quando o modelo previu que a frase pertence a classe "não gostei" e na verdade a frase pertence a categoria "gostei".

- **Falsos Positivos (False Positives - FP):** São os casos em que o modelo errou ao prever que uma instância pertence à classe positiva, quando na verdade a instância pertence à classe negativa. Por exemplo, quando o modelo previu que a frase pertence a classe "gostei" e na verdade a frase pertence a categoria "não gostei".

- **Verdadeiros Negativos (True Negatives - TN):** São os casos em que o modelo fez uma previsão correta, identificando corretamente uma instância como pertencente à classe negativa, e essa instância realmente pertence à classe negativa. Por exemplo, quando o modelo previu que a frase é do rotulo "não gostei" e de fato a frase pertence a essa categoria.

```python
# Importa funções para avaliação de modelos de classificação, como classification_report e confusion_matrix.
from sklearn.metrics import confusion_matrix

# Importa as bibliotecas seaborn e matplotlib.pyplot para visualização de dados.
import seaborn as sns
import matplotlib.pyplot as plt

# Previsões
y_test_predict = clf.predict(X_test_vectorized)

# Matriz de Confusão com Seaborn
confusion = confusion_matrix(y_test, y_test_predict)
labels = ['gostei', 'não gostei']

plt.figure(figsize=(8, 4))
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Previsto')
plt.ylabel('Verdadeiro')
plt.title('Matriz de Confusão')
plt.show()
```

#### 2.3.1 Métricas

Nessa etapa vamos submeter as frase do dataframe de teste e pedir para o modelo treinado predizer os sentimentos. Vamos comparar a predição com os sentimentos reais para calcular as métricas de presicion, de recall e f-measure.

- **Precisão (Precision):**

A precisão mede a proporção de previsões corretas (verdadeiros positivos) em relação ao total de previsões positivas feitas pelo modelo (verdadeiros positivos mais falsos positivos). A precisão é importante quando queremos minimizar os falsos positivos.

- **Revocação (Recall):**

A revocação mede a proporção de previsões corretas (verdadeiros positivos) em relação ao total de amostras que realmente pertencem à classe positiva (verdadeiros positivos mais falsos negativos). A revocação é relevante quando é fundamental identificar todas as instâncias positivas, minimizando assim os falsos negativos.

- **F1-Score (F-measure):**

O F1-Score é a média harmônica entre a precisão e a revocação. Ele combina ambas as métricas em uma única medida, dando mais peso à menor delas. O F1-Score é útil quando você deseja encontrar um equilíbrio entre precisão e revocação. Ele é especialmente útil quando as classes estão desequilibradas, pois evita que o desempenho seja dominado por uma das métricas.

- **Acurácia (Accuracy):**

A acurácia mede a proporção de previsões corretas em relação ao total de previsões feitas pelo modelo. A acurácia é uma métrica geral que indica o quão bem o modelo está performando no geral. No entanto, pode ser enganosa em cenários com classes desequilibradas, onde um alto número de amostras pertence a uma classe e poucas pertencem à outra.

```python
# Importa funções para avaliação de modelos de classificação, como classification_report e confusion_matrix.
from sklearn.metrics import classification_report

print('Relatório de Classificação:')
print(classification_report(y_test, y_test_predict))
```

#### 2.3.2 Possíveis problemas

- **Underfitting:**

Se o modelo tiver um desempenho ruim nos dados de treinamento (baixa precisão, baixo recall, alta taxa de erro, etc.) e também tiver um desempenho ruim nos dados de teste, é um sinal de underfitting. Isso indica que o modelo é muito simples para capturar a complexidade dos dados.


- **Overfitting:**

Alto desempenho nos dados de treinamento e baixo desempenho nos dados de teste podem ser um sinal clássico de overfitting. É quando o modelo tem um desempenho muito bom nos dados de treinamento, mas um desempenho significativamente pior nos dados de teste. O modelo "decorou" os dados de treinamento, mas não consegue generalizar bem para novos dados.

### 3.1 Perceptron

O perceptron é um **modelo linear de classificação binária** que requer que os dados sejam **linearmente separáveis** para funcionar adequadamente. Ele aprende por meio do cálculo de pesos e vieses e foi inspirado no funcionamento do neurônio humano. Proposto em 1943 por McCulloch e Pitt negritos e implementado em 1958 por Frank Rosenblatt, o perceptron foi inicialmente considerado um passo importante em direção à criação de um computador eletrônico com habilidades avançadas, como andar, falar, ver e ter consciência.

![deeplearning](https://drive.google.com/uc?id=1o4DN3KB8KvGOh-8kwzvI02KnsyR3_jFb)

Os pesos são atualizados com base no algoritmo de aprendizagem de descida gradiente (gradient descent learning algorithm). Ou seja, os pesos são atualizados de acordo com o feedback do resultado da função de ativação, que é uma função de degrau unitário. Segue a formula:

![deeplearning](https://drive.google.com/uc?id=1YiqaNT7iYyRWvJjHnYC7OB0vHVAU9Mrb)

Segue a implementação de uma classe com as principais funções de um Perceptron.

Link com detalhamento da implementação: https://vitalflux.com/perceptron-explained-using-python-example/amp/.


```python
import numpy as np

# Definição da classe Perceptron
class CustomPerceptron(object):

    # Construtor passando quantidade de iteração, estado aleatório e taxa de aprendizagem
    def __init__(self, n_iterations=100, random_state=1, learning_rate=0.01):
        self.n_iterations = n_iterations
        self.random_state = random_state
        self.learning_rate = learning_rate


    # Função de treinamento
    # 1. Os pesos são alterados baseado em cada exemplo de treinamento
    # 2. O aprendizado dos pesos continua por multiplas iterações
    # 3. A taxa de aprendizagem precisa ser definida
    def fit(self, X, y):
        rgen = np.random.RandomState(self.random_state)
        self.coef_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.errors_ = []
        for _ in range(self.n_iterations):
            errors = 0
            for xi, expected_value in zip(X, y):
                predicted_value = self.predict(xi)
                self.coef_[1:] = self.coef_[1:] + self.learning_rate * (expected_value - predicted_value) * xi
                self.coef_[0] = self.coef_[0] + self.learning_rate * (expected_value - predicted_value) * 1
                update = self.learning_rate * (expected_value - predicted_value)
                errors += int(update != 0.0)
            self.errors_.append(errors)

    # Função responsável por realizar a soma ponderada dos pesos
    def net_input(self, X):
        # Inicializa a soma ponderada com o peso de viés (coef_[0])
        weighted_sum = self.coef_[0]
        # Multiplica os valores de entrada pelos pesos correspondentes e soma
        weighted_sum += np.sum(X * self.coef_[1:])
        return weighted_sum

    # Método responsável por executar a função de ativação
    def activation_function(self, X):
        weighted_sum = self.net_input(X)
        return np.where(weighted_sum >= 0.0, 1, 0)

    # Função responsável por prever através ds função de ativação
    def predict(self, X):
        return self.activation_function(X)

    # Função que calcula a acurácia comparando o resultado esperado com o resultado previsto
    def score(self, X, y):
        misclassified_data_count = 0
        for xi, target in zip(X, y):
            output = self.predict(xi)
            if(target != output):
                misclassified_data_count += 1
        total_data_count = len(X)
        self.score_ = (total_data_count - misclassified_data_count)/total_data_count
        return self.score_
```

```python
import os

# Define o diretório onde estão localizados os arquivos de imagem.
data_dir = '/content/train'

# Inicializa contadores para contar o número de arquivos que começam com "dog" e "cat".
dog_count = 0
cat_count = 0

# Itera sobre os arquivos no diretório especificado.
for filename in os.listdir(data_dir):
    # Verifica se o nome do arquivo começa com "dog".
    if filename.startswith("dog"):
        dog_count += 1  # Incrementa o contador de arquivos "dog".
    # Verifica se o nome do arquivo começa com "cat".
    elif filename.startswith("cat"):
        cat_count += 1  # Incrementa o contador de arquivos "cat".

# Imprime a quantidade de arquivos que começam com "dog" e "cat".
print(f"Quantidade de 'dog': {dog_count}")
print(f"Quantidade de 'cat': {cat_count}")
```

```python
import tensorflow as tf
from tensorflow import keras

# Define o tamanho desejado para as imagens e o tamanho do lote para carregamento durante o treinamento.
image_size = (180, 180)
batch_size = 16

# Cria um conjunto de dados de treinamento usando imagens do diretório "PetImages".
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "PetImages",
    validation_split=0.2,  # Divide automaticamente 20% das imagens para validação.
    subset="training",      # Indica que este é o conjunto de treinamento.
    seed=1337,              # Define uma semente para reprodutibilidade.
    image_size=image_size,  # Define o tamanho desejado para as imagens.
    batch_size=batch_size   # Define o tamanho do lote para treinamento em mini lotes.
)

# Cria um conjunto de dados de validação usando imagens do diretório "PetImages".
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "PetImages",
    validation_split=0.2,  # Mesma divisão de validação usada no conjunto de treinamento.
    subset="validation",    # Indica que este é o conjunto de validação.
    seed=1337,              # Mesma semente usada para reprodutibilidade.
    image_size=image_size,  # Mesmo tamanho desejado para as imagens.
    batch_size=batch_size   # Mesmo tamanho do lote usado no conjunto de treinamento.
)
```

```python
# Cria uma sequência de transformações de aumento de dados usando TensorFlow e Keras.
data_augmentation = keras.Sequential(
    [
        # Camada que aplica uma virada horizontal aleatória nas imagens.
        tf.keras.layers.experimental.preprocessing.RandomFlip("horizontal"),

        # Camada que aplica uma rotação aleatória com ângulo máximo de 0.1 radianos.
        tf.keras.layers.experimental.preprocessing.RandomRotation(0.1),
    ]
)
```

##### **Modelagem**

O pré-carregamento (prefetching) é uma técnica que permite que os dados sejam carregados de forma eficiente em lotes antecipadamente durante o treinamento ou validação, enquanto o modelo está ocupado processando o lote atual. Isso ajuda a melhorar o desempenho do treinamento e validação, pois reduz o tempo de espera para carregar dados no momento em que são necessários, mantendo a CPU ou GPU ocupada com o processamento.

```python
# Prefetching (Pré-carregamento) é uma técnica usada para otimizar a velocidade de treinamento do modelo.

# Prefetch do conjunto de dados de treinamento (train_ds) para acelerar o treinamento.
train_ds = train_ds.prefetch(buffer_size=32)

# Prefetch do conjunto de dados de validação (val_ds) para acelerar o processo de validação.
val_ds = val_ds.prefetch(buffer_size=32)
```

```python
# Função para criar o modelo de rede neural.
def make_model(input_shape, num_classes):
    # Define a camada de entrada com a forma especificada (input_shape).
    inputs = keras.Input(shape=input_shape)

    # Bloco de aumento de dados de imagem.
    x = data_augmentation(inputs)

    # Bloco de pré-processamento de entrada (normalização de pixels).
    x = tf.keras.layers.experimental.preprocessing.Rescaling(1.0 / 255)(x)

    # Camada de convolução 2D com 32 filtros, tamanho do filtro 3x3, passo de 2 e valid padding.
    x = tf.keras.layers.Conv2D(32, 3, strides=2, padding="valid")(x)
    x = tf.keras.layers.BatchNormalization()(x)  # Normalização por lotes.
    x = tf.keras.layers.Activation("relu")(x)    # Função de ativação ReLU.
    x = tf.keras.layers.MaxPooling2D(2, strides=2, padding="valid")(x)  # Camada de pooling.

    # Camada de convolução 2D com 64 filtros e tamanho do filtro 3x3.
    x = tf.keras.layers.Conv2D(64, 3, padding="valid")(x)
    x = tf.keras.layers.BatchNormalization()(x)  # Normalização por lotes.
    x = tf.keras.layers.Activation("relu")(x)    # Função de ativação ReLU.
    x = tf.keras.layers.MaxPooling2D(2, strides=2, padding="valid")(x)  # Camada de pooling.

    # Camada de convolução 2D com 128 filtros e tamanho do filtro 3x3.
    x = tf.keras.layers.Conv2D(128, 3, padding="valid")(x)
    x = tf.keras.layers.BatchNormalization()(x)  # Normalização por lotes.
    x = tf.keras.layers.Activation("relu")(x)    # Função de ativação ReLU.
    x = tf.keras.layers.MaxPooling2D(2, strides=2, padding="valid")(x)  # Camada de pooling.

    # Achatamento das características extraídas.
    x = tf.keras.layers.Flatten()(x)

    # Camada oculta (hidden layer) com 128 unidades e ativação ReLU.
    x = tf.keras.layers.Dense(128, activation='relu')(x)

    # Camada de saída com uma única unidade e ativação sigmoid para classificação binária.
    outputs = tf.keras.layers.Dense(1, activation="sigmoid")(x)

    # Cria o modelo final com as camadas definidas.
    return keras.Model(inputs, outputs)

# Criação do modelo com as dimensões de entrada especificadas e 2 classes de saída.
model = make_model(input_shape=image_size + (3,), num_classes=2)

# Mostra o resumo do modelo
model.summary()
```

##### **Treinamento**

```python
# Define o número de épocas (ciclos de treinamento) desejado.
epochs = 20

# Compilação do modelo com otimizador Adam, função de perda "binary_crossentropy" e métrica "accuracy".
model.compile(
    optimizer=keras.optimizers.Adam(1e-3),      # Otimizador Adam com taxa de aprendizado de 0,001.
    loss="binary_crossentropy",                 # Função de perda para classificação binária.
    metrics=["accuracy"]                        # Métrica de acurácia para avaliar o desempenho do modelo.
)

# Treinamento do modelo usando os dados de treinamento (train_ds) e validação (val_ds).
history_1 = model.fit(
    train_ds,                                   # Conjunto de treinamento.
    epochs=epochs,                             # Número de épocas.
    validation_data=val_ds,                    # Dados de validação.
    verbose=1,                                 # Exibe informações durante o treinamento.
)
```

##### **Acurácia e perda**

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")

# Crie uma figura com duas subtramas (um para perda e outro para precisão).
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6))

# Gráfico de linha para perda (loss).
sns.lineplot(x=range(1, epochs+1), y=history_1.history['loss'], color='b', label="Training loss", ax=ax1)
sns.lineplot(x=range(1, epochs+1), y=history_1.history['val_loss'], color='r', label="Validation loss", ax=ax1)
ax1.set_xticks(range(1, epochs+1))
ax1.set_yticks(np.arange(0, 1, 0.1))
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.set_title('Training and Validation Loss')

# Gráfico de linha para precisão (accuracy).
sns.lineplot(x=range(1, epochs+1), y=history_1.history['accuracy'], color='b', label="Training accuracy", ax=ax2)
sns.lineplot(x=range(1, epochs+1), y=history_1.history['val_accuracy'], color='r', label="Validation accuracy", ax=ax2)
ax2.set_xticks(range(1, epochs+1))
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.set_title('Training and Validation Accuracy')

# Adicione uma legenda.
legend = plt.legend(loc='best', shadow=True)

# Ajuste o layout e exiba o gráfico.
plt.tight_layout()
plt.show()
```
